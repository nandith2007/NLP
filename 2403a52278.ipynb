```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
text = "Natural Language Processing is interesting."

# NLTK tokenization
import nltk
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize
print(word_tokenize(text))

# spaCy processing
doc = nlp(text)
for token in doc:
    print(token.text, token.pos_)
```

```python
text = """
Artificial Intelligence is transforming industries across the world.
From healthcare to education, AI-powered systems are improving efficiency,
reducing costs, and enabling smarter decision-making.
"""

# Convert text to lowercase
lower_text = text.lower()

# Count number of words
word_count = len(text.split())

print("Lowercase Text:\n", lower_text)
print("\nNumber of words:", word_count)
```
